```python
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size= 0.2, random_state=5)


```

```python
model = SGDClassifier(random_state=0)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
# transformer
scaler = StandardScaler()
X_train_transformer = scaler.fit_transform(X_train)

# Estimator
classifier = SGDClassifier(random_state=0)
classifier.fit(X_train_transformer,y_train)

# test 
X_test_transformer = scaler.fit_transform(X_test)
y_pred = classifier.predict(X_test_transformer)
classifier.score(X_test_transformer,y_test)
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipeline = make_pipeline(StandardScaler(),
                         SGDClassifier(random_state = 0))

pipeline.fit(X_train,y_train)
pipeline.predict(X_test)
pipeline.score(X_test,y_test)                         
```
